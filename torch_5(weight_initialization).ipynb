```python
# case 1: inititalising all the weights and bias with 0 and avtifatcion sigmoid

class BinarClass(t.nn.Module):
  def __init__(self,n_features):
    super().__init__()

    self.ann=t.nn.Sequential(
        t.nn.Linear(n_features,6),
        t.nn.Sigmoid(),
        t.nn.Linear(6,4),
        t.nn.Sigmoid(),
        t.nn.Linear(4,3),
        t.nn.Sigmoid(),
        t.nn.Linear(3,1))

  def forward(self,data):
    return self.ann(data)

model=BinarClass(2)

def weight_intit(x):
  if isinstance(x,t.nn.Linear):
    t.nn.init.zeros_(x.weight)
    t.nn.init.zeros_(x.bias)

model.apply(weight_intit)

list(model.parameters())
```

```python


# define the loss

loss=t.nn.BCEWithLogitsLoss()

# define the optimiser
opt=t.optim.SGD(model.parameters(),lr=0.9)

epochs=300
sigmoid=t.nn.Sigmoid()

train_loss_data=[]
cv_loss_data=[]

train_acc_data=[]
cv_acc_data=[]
for iter in range(epochs):
  model.train()

  predi_t=model(x_trains)

  prob_output_t=sigmoid(predi_t)

  pr_labeltrain=[]
  for prob in prob_output_t.squeeze():
    if prob>=0.5:
      pr_labeltrain.append(1)
    else:
      pr_labeltrain.append(0)

  train_acc_data.append(accuracy_score(y_train,pr_labeltrain))

  train_loss=loss(predi_t.squeeze(),y_train)

  opt.zero_grad()

  train_loss.backward()

  opt.step()

  model.eval()

  with t.no_grad():
    predi_cv=model(x_cvs)
    prob_output_cv=sigmoid(predi_cv)

    pr_labelcv=[]
    for prob in prob_output_cv.squeeze():
      if prob>=0.5:
        pr_labelcv.append(1)
      else:
        pr_labelcv.append(0)

    cv_acc_data.append(accuracy_score(y_cv,pr_labelcv))

    cv_loss=loss(predi_cv.squeeze(),y_cv)

  train_loss_data.append(train_loss.detach().numpy())
  cv_loss_data.append(cv_loss.numpy())

  print("epochs :{}, train_loss :{}, cv_loss :{} // train_acc :{}, cv_acc :{}".format(iter+1,train_loss,cv_loss,accuracy_score(y_train,pr_labeltrain),accuracy_score(y_cv,pr_labelcv)))

```

```python
# cas2 2 init - with a constant value sigmoid,relu,tanh
# case1 b acitvation function == relu

# case 1: inititalising all the weights and bias with 0 and avtifatcion sigmoid

class BinarClass(t.nn.Module):
  def __init__(self,n_features):
    super().__init__()

    self.ann=t.nn.Sequential(
        t.nn.Linear(n_features,6),
        t.nn.ReLU(),
        t.nn.Linear(6,4),
        t.nn.ReLU(),
        t.nn.Linear(4,3),
        t.nn.ReLU(),
        t.nn.Linear(3,1))

  def forward(self,data):
    return self.ann(data)

model=BinarClass(2)

def weight_intit(x):
  if isinstance(x,t.nn.Linear):
    t.nn.init.constant_(x.weight,1)
    t.nn.init.constant_(x.bias,1)

model.apply(weight_intit)

print("before trainning:- ",list(model.parameters()))



# define the loss

loss=t.nn.BCEWithLogitsLoss()

# define the optimiser
opt=t.optim.SGD(model.parameters(),lr=0.9)

epochs=100
sigmoid=t.nn.Sigmoid()

train_loss_data=[]
cv_loss_data=[]

train_acc_data=[]
cv_acc_data=[]
for iter in range(epochs):
  model.train()

  predi_t=model(x_trains)

  prob_output_t=sigmoid(predi_t)

  pr_labeltrain=[]
  for prob in prob_output_t.squeeze():
    if prob>=0.5:
      pr_labeltrain.append(1)
    else:
      pr_labeltrain.append(0)

  train_acc_data.append(accuracy_score(y_train,pr_labeltrain))

  train_loss=loss(predi_t.squeeze(),y_train)

  opt.zero_grad()

  train_loss.backward()

  opt.step()

  model.eval()

  with t.no_grad():
    predi_cv=model(x_cvs)
    prob_output_cv=sigmoid(predi_cv)

    pr_labelcv=[]
    for prob in prob_output_cv.squeeze():
      if prob>=0.5:
        pr_labelcv.append(1)
      else:
        pr_labelcv.append(0)

    cv_acc_data.append(accuracy_score(y_cv,pr_labelcv))

    cv_loss=loss(predi_cv.squeeze(),y_cv)

  train_loss_data.append(train_loss.detach().numpy())
  cv_loss_data.append(cv_loss.numpy())

print("after trainning:- " ,list(model.parameters()))
```
